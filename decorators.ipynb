```python
def add(n):
    print(f"Adding started")
    res= 1
    for i in range(1, n + 1):
        res += i
    print(f"Adding is over")
    return res

def fact(n):
    print(f"Fact started")
    res= 1
    for i in range(1, n + 1):
        res *= i
    print(f"Fact is over")
    return res

f= fact(5)
print(f)
s= add(5)
print(s)
```

```python
## Hard values (no arguments) are passed to the functions, so they will be executed in the main thread.

def logger(func):
    def inner():
        print(f"Function {func.__name__} started")
        result = func()
        print(f"Function {func.__name__} ended")
        return result
    return inner

@logger
def add():
    n= 5
    res= 1
    for i in range(1, n + 1):
        res += i
    return res
    
@logger 
def fact():
    n= 5
    res= 1
    for i in range(1, n + 1):
        res *= i
    return res

f= fact()
print(f)
s= add()
print(s)
```

```python
def logger(func):
    def inner(*args, **kwargs):
        print(f"Function {func.__name__} started")
        result = func(*args, **kwargs)
        print(f"Function {func.__name__} ended")
        return result
    return inner

@logger
def add(n):
    res= 1
    for i in range(1, n + 1):
        res += i
    return res
    
@logger 
def fact(n):
    res= 1
    for i in range(1, n + 1):
        res *= i
    return res

f= fact(5)
print(f)
s= add(5)
print(s)
```

```python
import time

def time_logger(func):
    def inner(*args, **kwargs):
        print(f"At {time.time()}, Function {func.__name__} started")
        result = func(*args, **kwargs)
        print(f"At {time.time()}, Function {func.__name__} ended")
        return result
    return inner

@time_logger
def add(n):
    res= 1
    for i in range(1, n + 1):
        res += i
    return res
    
@time_logger
def fact(n):
    res= 1
    for i in range(1, n + 1):
        res *= i
    return res

f= fact(5)
print(f)
s= add(5)
print(s)
```
