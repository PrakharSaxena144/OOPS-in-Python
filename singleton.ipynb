```python
class FamilyMember:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        return f"Hello, my name is {self.name}."

fm1 = FamilyMember("Alice")
print(fm1.introduce())
```

```python
from abc import ABC, abstractmethod

class FamilyMember(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def introduce(self):
        pass

class KidMember(FamilyMember):
    def introduce(self):
        return f"Hi, I'm {self.name}, and I'm a kid."
```
